```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## Time for 100 epochs
nbins = [2,4,8,10,12,14] # batchsize 
ipu = [372.02, 350.73, 445.53, 406.18, 407.61, 438.58]
gpu = [2129.34, 1079.03, 597.34, 509.71, 500.70, 454.67]
```

```python
plt.plot(nbins, gpu, '-', label = 'GPU')
plt.plot(nbins, ipu, '-', label = 'IPU')
plt.legend()
plt.xlabel('Batchsize')
plt.ylabel('Time (s)')
plt.title('Training on 100 epochs')
```
