# Test

```python
n = 10
```

```python
test_data = np.zeros((n*batch_size, 64, 64, 6))
test_labels = np.zeros((n*batch_size, 2))

for i in range (n):
    test = test_generator.__getitem__(2)
    test_data[i*batch_size: (i+1)*batch_size] = test[0]
    test_labels[i*batch_size: (i+1)*batch_size] = test[1]
```

```python
out = net(test_data)
```

```python
out_mean = np.zeros((n*batch_size,final_dim))
out_stddev = np.zeros((n*batch_size,final_dim))

for i in range (final_dim):
    out_mean[:,i] = out.mean().numpy()[:,i]
    out_stddev[:,i] = out.stddev().numpy()[:,i]                          
```

```python
sns.distplot(out_mean[:,0], bins = 20)
sns.distplot(test_labels[:,0], bins = 20)
```

```python
sns.distplot(out_stddev[:,0])
sns.distplot(out_stddev[:,1])
print("median dx: "+str(np.median(out_stddev[:,0])), "median dy: "+str(np.median(out_stddev[:,1])))
print("mean dx: "+str(np.mean(out_stddev[:,0])), "mean dy: "+str(np.mean(out_stddev[:,1])))
```

### One sampling for 1000 images (ten batch)

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

for i in range (2):
    axes[i].plot(test_labels[:, i], out_mean[:,i], '.', label = 'mean')
    axes[i].plot(test_labels[:, i], out_mean[:,i]+ 2*out_stddev[:,i], '+', label = 'mean + 2stddev')
    axes[i].plot(test_labels[:, i], out_mean[:,i]- 2*out_stddev[:,i], '+', label = 'mean - 2stddev')
    x = np.linspace(-4,4)
    axes[i].plot(x, x)
    axes[i].legend()
    #axes[0].set_ylim(-0.5,6.5)
    if i == 0:
        axes[i].set_title('$shift_x$')
    else:
        axes[i].set_title('$shift_y$')
```

### N samplings for 100 images (one batch)

```python
N = 100
means_e1 = np.zeros((N,100))
stddevs_e1 = np.zeros((N,100))
means_e2 = np.zeros((N,100))
stddevs_e2 = np.zeros((N,100))

outs = np.zeros((N,100))
for i in range (N):
    out = net(test_data[:100])
    outs[i] = out[:,0].numpy()
    means_e1[i] = out.mean().numpy()[:,0]
    stddevs_e1[i] = out.stddev().numpy()[:,0]
    means_e2[i] = out.mean().numpy()[:,1]
    stddevs_e2[i] = out.stddev().numpy()[:,1]
```

```python
image_nb =40
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
x = np.linspace(-5,5, 1000)


axes[0].plot(x, (1/(stddevs_e1[0,image_nb]*np.sqrt(2*np.pi))*np.exp(-np.square(x-means_e1[0,image_nb])/(2*stddevs_e1[0,image_nb]**2))), color = 'grey', alpha = 0.6, label = '$shift_x$')
axes[0].axvline(test_labels[image_nb,0], color = 'r', label='true $shift_x`$')

axes[1].plot(x, (1/(stddevs_e2[i,image_nb]*np.sqrt(2*np.pi))*np.exp(-np.square(x-means_e2[i,image_nb])/(2*stddevs_e2[i,image_nb]**2))), color = 'b', alpha = 0.6, label= '$shift_y$')
axes[1].axvline(test_labels[image_nb,1], color = 'green', label='true $shift_y$')

axes[2].imshow(test_data[image_nb,:,:,2])
axes[0].legend()
axes[1].legend()

min_shift_x = np.min(means_e1[0,image_nb]-4*stddevs_e1[0,image_nb])
min_shift_y = np.min(means_e2[0,image_nb]-4*stddevs_e2[0,image_nb])
max_shift_x = np.max(means_e1[0,image_nb]+4*stddevs_e1[0,image_nb])
max_shift_y = np.max(means_e2[0,image_nb]+4*stddevs_e2[0,image_nb])
axes[0].set_xlim(min_shift_x-0.2, max_shift_x+0.2)
axes[1].set_xlim(min_shift_y-0.2, max_shift_y+0.2)
```
