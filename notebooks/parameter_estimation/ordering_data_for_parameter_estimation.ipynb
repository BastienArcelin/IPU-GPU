```python
import sys
import os
sys.path.insert(0,'../../scripts/tools_for_VAE/')
import tools_for_VAE.layers as layers
from tools_for_VAE import utils, vae_functions, generator, model
```

```python
images_dir = '/sps/lsst/users/barcelin/data/TFP/GalSim_COSMOS/blended_galaxies/random/'
root = 'galaxies_blended_20191024'
data_dir = images_dir + 'test/'
list_of_samples = [x for x in utils.listdir_fullpath(os.path.join(images_dir,'test')) if x.endswith('.npy')]
```

```python
list_of_samples
```

```python
for k in range (len(list_of_samples)):
    
    dfs = []

    dfs.append(pd.read_csv(os.path.join(data_dir, root+'_'+str(k)+'_data.csv')))
    df = dfs[0]

    for df2 in dfs[1:]:
        df = df.append(df2)
        
        
    keys = []
    for i in range (4):
        keys = keys+['redshift_'+str(i), 'e1_'+str(i), 'e2_'+str(i), 'mag_'+str(i)]
    df_new = pd.DataFrame(index=np.arange(len(df)), columns=keys)
    
    for i in range(len(df)):
        if np.array(df['nb_blended_gal'][i]>1):
            df[df==10]=30
            _idx = np.argmin((df['mag_1'][i], df['mag_2'][i], df['mag_3'][i]))
            arr_mag = np.delete((df['mag_1'][i], df['mag_2'][i], df['mag_3'][i]), _idx)
            _idx_2 = np.argmin(arr_mag)

            arr_e1= np.delete((df['e1_1'][i], df['e1_2'][i], df['e1_3'][i]), _idx)
            arr_e2 = np.delete((df['e2_1'][i], df['e2_2'][i], df['e2_3'][i]), _idx)
            arr_z = np.delete((df['redshift_1'][i], df['redshift_2'][i], df['redshift_3'][i]), _idx)

            df_new['e1_0'][i] = df['e1_0'][i]
            df_new['e2_0'][i] = df['e2_0'][i]
            df_new['redshift_0'][i] = df['redshift_0'][i]
            df_new['mag_0'][i] = df['mag_0'][i]

            df_new['e1_1'][i] = np.array(df['e1_'+str(_idx+1)][i])
            df_new['e2_1'][i] = np.array(df['e2_'+str(_idx+1)][i])
            df_new['redshift_1'][i] = np.array(df['redshift_'+str(_idx+1)][i])
            df_new['mag_1'][i] = np.array(df['mag_'+str(_idx+1)][i])

            df_new['e1_2'][i] = arr_e1[_idx_2]
            df_new['e2_2'][i] = arr_e2[_idx_2]
            df_new['redshift_2'][i] = arr_z[_idx_2]
            df_new['mag_2'][i] = arr_mag[_idx_2]

            df_new['e1_3'][i] = np.delete(arr_e1, _idx_2)[0]
            df_new['e2_3'][i] = np.delete(arr_e2, _idx_2)[0]
            df_new['redshift_3'][i] = np.delete(arr_z, _idx_2)[0]
            df_new['mag_3'][i] = np.delete(arr_mag, _idx_2)[0]
        else:
            df_new['e1_0'][i] = df['e1_0'][i]
            df_new['e2_0'][i] = df['e2_0'][i]
            df_new['redshift_0'][i] = df['redshift_0'][i]
            df_new['mag_0'][i] = df['mag_0'][i]

            df_new['e1_1'][i] = 0
            df_new['e2_1'][i] = 0
            df_new['redshift_1'][i] = 0
            df_new['mag_1'][i] = 0

            df_new['e1_2'][i] = 0
            df_new['e2_2'][i] = 0
            df_new['redshift_2'][i] = 0
            df_new['mag_2'][i] = 0

            df_new['e1_3'][i] = 0
            df_new['e2_3'][i] = 0
            df_new['redshift_3'][i] = 0
            df_new['mag_3'][i] = 0    
    
    df_new.to_csv(os.path.join(data_dir, root+'_'+str(k)+'_data_classified.csv'))
    print(k)
```
