```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

Plot function

```python
def plot_rgb(gal, bands=[5,6,7], ax=None, band_first=True, zoom=1.5, shifts=None):
    '''
    Return plot in rgb of image
    '''
    if ax is None:
        ax = plt.subplot()
    if band_first:
        tr = [1,2,0]
    else:
        tr = [0,1,2]
    
    imsize = float(gal.shape[1]) / 2.
    ax.imshow(np.clip(gal[:,:,:].transpose(tr)[:,:,bands], a_min=0.0, a_max=1.), origin='center')
    if shifts is not None:
        for (x,y) in shifts:
            ax.scatter(y, x,  marker='+', c='b')
        #ax.scatter(0., 0., marker='+', c='b')
    #ax.set_xlim(-imsize/zoom,imsize/zoom)
    #ax.set_ylim(-imsize/zoom,imsize/zoom)
    ax.axis('off')
```
